```python
import os, tarfile, random
import numpy as np
import nibabel as nib
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import torch.nn.functional as F
from torch.optim.lr_scheduler import CosineAnnealingWarmRestarts
from torch.cuda.amp import autocast, GradScaler

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

extract_path = '/kaggle/working/extracted_data'
if not os.path.exists(extract_path):
    os.makedirs(extract_path, exist_ok=True)
    with tarfile.open('/kaggle/input/brats-2021-task1/BraTS2021_Training_Data.tar') as tar:
        tar.extractall(path=extract_path)


def augment(img, mask):
    if random.random() > 0.5:
        img = torch.flip(img, [2])
        mask = torch.flip(mask, [2])
    
    if random.random() > 0.5:
        img = torch.flip(img, [1])
        mask = torch.flip(mask, [1])
    
    k = random.randint(0, 3)
    if k > 0:
        img = torch.rot90(img, k, [1, 2])
        mask = torch.rot90(mask, k, [1, 2])
    
    if random.random() > 0.5:
        h, w = img.shape[1], img.shape[2]
        crop_size = int(random.uniform(0.85, 0.95) * h)
        top = random.randint(0, h - crop_size)
        left = random.randint(0, w - crop_size)
        img = img[:, top:top+crop_size, left:left+crop_size]
        mask = mask[:, top:top+crop_size, left:left+crop_size]
        img = F.interpolate(img.unsqueeze(0), size=(h, w), mode='bilinear', align_corners=False).squeeze(0)
        mask = F.interpolate(mask.unsqueeze(0), size=(h, w), mode='nearest').squeeze(0)
    
    if random.random() > 0.5:
        img = img * random.uniform(0.85, 1.15) + random.uniform(-0.15, 0.15)
    
    if random.random() > 0.7:
        noise = torch.randn_like(img) * 0.05
        img = img + noise
    
    return img, mask


class BraTSDataset(Dataset):
    def __init__(self, patient_ids, root_dir, augment_data=False, slices_per_patient=3):
        self.patient_ids = patient_ids
        self.root_dir = root_dir
        self.augment_data = augment_data
        self.slices_per_patient = slices_per_patient

    def __len__(self):
        return len(self.patient_ids) * self.slices_per_patient

    def __getitem__(self, idx):
        pid = self.patient_ids[idx // self.slices_per_patient]
        img_path = os.path.join(self.root_dir, pid, f"{pid}_flair.nii.gz")
        mask_path = os.path.join(self.root_dir, pid, f"{pid}_seg.nii.gz")
        
        img_vol = nib.load(img_path).get_fdata()
        mask_vol = nib.load(mask_path).get_fdata()

        tumor_slices = np.where(mask_vol.sum(axis=(0, 1)) > 50)[0]
        if len(tumor_slices) > 0:
            z = random.choice(tumor_slices)
        else:
            z = img_vol.shape[2] // 2

        img = img_vol[:, :, z]
        mask = mask_vol[:, :, z]

        if img.max() > 0:
            p1, p99 = np.percentile(img[img > 0], [1, 99])
            img = np.clip(img, p1, p99)
            img = (img - img.mean()) / (img.std() + 1e-8)
        
        mask = (mask > 0).astype(np.float32)

        img = torch.tensor(img).unsqueeze(0).float()
        mask = torch.tensor(mask).unsqueeze(0).float()

        if self.augment_data:
            img, mask = augment(img, mask)

        return img, mask


class AttentionBlock(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__()
        self.W_g = nn.Sequential(
            nn.Conv2d(F_g, F_int, 1, bias=False),
            nn.BatchNorm2d(F_int)
        )
        self.W_x = nn.Sequential(
            nn.Conv2d(F_l, F_int, 1, bias=False),
            nn.BatchNorm2d(F_int)
        )
        self.psi = nn.Sequential(
            nn.Conv2d(F_int, 1, 1, bias=False),
            nn.BatchNorm2d(1),
            nn.Sigmoid()
        )
        self.relu = nn.ReLU(inplace=True)

    def forward(self, g, x):
        g1 = self.W_g(g)
        x1 = self.W_x(x)
        psi = self.relu(g1 + x1)
        psi = self.psi(psi)
        return x * psi


class ResidualBlock(nn.Module):
    def __init__(self, in_c, out_c, dropout=0.1):
        super().__init__()
        self.conv1 = nn.Conv2d(in_c, out_c, 3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_c)
        self.conv2 = nn.Conv2d(out_c, out_c, 3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_c)
        self.relu = nn.ReLU(inplace=True)
        self.dropout = nn.Dropout2d(dropout) if dropout > 0 else nn.Identity()
        
        self.skip = nn.Sequential(
            nn.Conv2d(in_c, out_c, 1, bias=False),
            nn.BatchNorm2d(out_c)
        ) if in_c != out_c else nn.Identity()

    def forward(self, x):
        residual = self.skip(x)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.dropout(x)
        x = self.bn2(self.conv2(x))
        return self.relu(x + residual)


class ASPP(nn.Module):
    def __init__(self, in_c, out_c):
        super().__init__()
        self.conv1 = nn.Conv2d(in_c, out_c, 1, bias=False)
        self.conv2 = nn.Conv2d(in_c, out_c, 3, padding=6, dilation=6, bias=False)
        self.conv3 = nn.Conv2d(in_c, out_c, 3, padding=12, dilation=12, bias=False)
        self.conv4 = nn.Conv2d(in_c, out_c, 3, padding=18, dilation=18, bias=False)
        self.global_pool = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Conv2d(in_c, out_c, 1, bias=False)
        )
        self.out = nn.Conv2d(out_c * 5, out_c, 1, bias=False)
        self.bn = nn.BatchNorm2d(out_c)
        self.relu = nn.ReLU(inplace=True)

    def forward(self, x):
        size = x.shape[2:]
        x1 = self.conv1(x)
        x2 = self.conv2(x)
        x3 = self.conv3(x)
        x4 = self.conv4(x)
        x5 = F.interpolate(self.global_pool(x), size=size, mode='bilinear', align_corners=False)
        x = torch.cat([x1, x2, x3, x4, x5], dim=1)
        return self.relu(self.bn(self.out(x)))


class ImprovedUNet(nn.Module):
    def __init__(self):
        super().__init__()

        self.enc1 = ResidualBlock(1, 64, dropout=0.1)
        self.enc2 = ResidualBlock(64, 128, dropout=0.1)
        self.enc3 = ResidualBlock(128, 256, dropout=0.1)
        self.enc4 = ResidualBlock(256, 512, dropout=0.1)
        self.pool = nn.MaxPool2d(2)

        self.bottleneck = ASPP(512, 1024)

        self.up4 = nn.ConvTranspose2d(1024, 512, 2, 2)
        self.att4 = AttentionBlock(512, 512, 256)
        self.dec4 = ResidualBlock(1024, 512, dropout=0.1)

        self.up3 = nn.ConvTranspose2d(512, 256, 2, 2)
        self.att3 = AttentionBlock(256, 256, 128)
        self.dec3 = ResidualBlock(512, 256, dropout=0.1)

        self.up2 = nn.ConvTranspose2d(256, 128, 2, 2)
        self.att2 = AttentionBlock(128, 128, 64)
        self.dec2 = ResidualBlock(256, 128, dropout=0.1)

        self.up1 = nn.ConvTranspose2d(128, 64, 2, 2)
        self.att1 = AttentionBlock(64, 64, 32)
        self.dec1 = ResidualBlock(128, 64, dropout=0.1)

        self.out = nn.Conv2d(64, 1, 1)

    def forward(self, x):
        s1 = self.enc1(x)
        s2 = self.enc2(self.pool(s1))
        s3 = self.enc3(self.pool(s2))
        s4 = self.enc4(self.pool(s3))
        
        b = self.bottleneck(self.pool(s4))

        d4 = self.up4(b)
        s4 = self.att4(d4, s4)
        d4 = self.dec4(torch.cat([d4, s4], dim=1))

        d3 = self.up3(d4)
        s3 = self.att3(d3, s3)
        d3 = self.dec3(torch.cat([d3, s3], dim=1))

        d2 = self.up2(d3)
        s2 = self.att2(d2, s2)
        d2 = self.dec2(torch.cat([d2, s2], dim=1))

        d1 = self.up1(d2)
        s1 = self.att1(d1, s1)
        d1 = self.dec1(torch.cat([d1, s1], dim=1))

        return torch.sigmoid(self.out(d1))


class DiceLoss(nn.Module):
    def forward(self, pred, target):
        smooth = 1.0
        pred = pred.view(pred.size(0), -1)
        target = target.view(target.size(0), -1)

        intersection = (pred * target).sum(dim=1)
        dice = (2 * intersection + smooth) / (pred.sum(dim=1) + target.sum(dim=1) + smooth)
        return 1 - dice.mean()


class TverskyLoss(nn.Module):
    def __init__(self, alpha=0.7, beta=0.3):
        super().__init__()
        self.alpha = alpha
        self.beta = beta

    def forward(self, pred, target):
        smooth = 1.0
        pred = pred.view(pred.size(0), -1)
        target = target.view(target.size(0), -1)

        tp = (pred * target).sum(dim=1)
        fp = (pred * (1 - target)).sum(dim=1)
        fn = ((1 - pred) * target).sum(dim=1)

        tversky = (tp + smooth) / (tp + self.alpha * fp + self.beta * fn + smooth)
        return 1 - tversky.mean()


class FocalTverskyLoss(nn.Module):
    def __init__(self, alpha=0.7, beta=0.3, gamma=1.5):
        super().__init__()
        self.tversky = TverskyLoss(alpha, beta)
        self.gamma = gamma

    def forward(self, pred, target):
        tversky_loss = self.tversky(pred, target)
        return torch.pow(tversky_loss, self.gamma)


dice_loss = DiceLoss()
focal_tversky = FocalTverskyLoss()


def combined_loss(pred, target):
    return dice_loss(pred, target) + focal_tversky(pred, target)


def compute_metrics(pred, target):
    pred = (pred > 0.5).float()

    if target.sum() == 0:
        return None, None

    inter = (pred * target).sum()
    union = pred.sum() + target.sum()

    dice = (2 * inter) / (union + 1e-6)
    iou = inter / (union - inter + 1e-6)

    return dice.item(), iou.item()


all_ids = [d for d in os.listdir(extract_path) if os.path.isdir(os.path.join(extract_path, d))]
train_ids, val_ids = train_test_split(all_ids[:500], test_size=0.12, random_state=42)

print(f"Training samples: {len(train_ids)}, Validation samples: {len(val_ids)}")

train_loader = DataLoader(
    BraTSDataset(train_ids, extract_path, augment_data=True, slices_per_patient=3), 
    batch_size=12, 
    shuffle=True, 
    num_workers=2,
    pin_memory=True
)
val_loader = DataLoader(
    BraTSDataset(val_ids, extract_path, augment_data=False, slices_per_patient=1), 
    batch_size=4, 
    num_workers=2,
    pin_memory=True
)

model = ImprovedUNet().to(device)
optimizer = torch.optim.AdamW(model.parameters(), lr=2e-4, weight_decay=1e-4)
scheduler = CosineAnnealingWarmRestarts(optimizer, T_0=10, T_mult=2)
scaler = GradScaler()

print("\nTraining started...\n")

best_dice = 0
patience_counter = 0
history = {'train_loss': [], 'val_dice': [], 'val_iou': []}

for epoch in range(50):
    model.train()
    train_loss = 0

    for imgs, masks in train_loader:
        imgs, masks = imgs.to(device), masks.to(device)
        
        optimizer.zero_grad()
        
        with autocast():
            pred = model(imgs)
            loss = combined_loss(pred, masks)
        
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()
        
        train_loss += loss.item()

    model.eval()
    dice_scores, iou_scores = [], []

    with torch.no_grad():
        for i, (imgs, masks) in enumerate(val_loader):
            imgs, masks = imgs.to(device), masks.to(device)
            
            with autocast():
                preds = model(imgs)

            for j in range(imgs.size(0)):
                d, iou = compute_metrics(preds[j:j+1], masks[j:j+1])
                if d is not None:
                    dice_scores.append(d)
                    iou_scores.append(iou)

            if i == 0 and epoch % 5 == 0:
                plt.figure(figsize=(15, 3))
                
                plt.subplot(1, 4, 1)
                plt.imshow(imgs[0, 0].cpu(), cmap='gray')
                plt.title("FLAIR Input")
                plt.axis('off')

                plt.subplot(1, 4, 2)
                plt.imshow(masks[0, 0].cpu(), cmap='jet', alpha=0.6)
                plt.title("Ground Truth")
                plt.axis('off')

                plt.subplot(1, 4, 3)
                plt.imshow(preds[0, 0].cpu() > 0.5, cmap='jet', alpha=0.6)
                plt.title("Prediction")
                plt.axis('off')

                plt.subplot(1, 4, 4)
                overlay = imgs[0, 0].cpu().numpy()
                pred_mask = (preds[0, 0].cpu() > 0.5).numpy()
                plt.imshow(overlay, cmap='gray')
                plt.imshow(pred_mask, cmap='jet', alpha=0.4)
                plt.title("Overlay")
                plt.axis('off')
                
                plt.tight_layout()
                plt.show()

    avg_dice = np.mean(dice_scores)
    avg_iou = np.mean(iou_scores)
    avg_loss = train_loss / len(train_loader)
    
    history['train_loss'].append(avg_loss)
    history['val_dice'].append(avg_dice)
    history['val_iou'].append(avg_iou)
    
    print(
        f"Epoch {epoch+1:2d} | "
        f"Loss: {avg_loss:.4f} | "
        f"Dice: {avg_dice:.4f} ({avg_dice*100:.1f}%) | "
        f"IoU: {avg_iou:.4f} ({avg_iou*100:.1f}%) | "
        f"LR: {optimizer.param_groups[0]['lr']:.6f}"
    )

    scheduler.step()

    if avg_dice > best_dice:
        best_dice = avg_dice
        patience_counter = 0
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'dice': best_dice,
        }, 'best_model.pth')
        print(f"New best! Dice: {best_dice:.4f} ({best_dice*100:.1f}%)")
    else:
        patience_counter += 1
        if patience_counter >= 12:
            print(f"\nEarly stopping at epoch {epoch+1}")
            break

print(f"\nTraining complete!")
print(f"Best Dice: {best_dice:.4f} ({best_dice*100:.1f}%)")
print(f"Best IoU: {max(history['val_iou']):.4f} ({max(history['val_iou'])*100:.1f}%)")

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.plot(history['train_loss'])
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(history['val_dice'])
plt.title('Validation Dice Score')
plt.xlabel('Epoch')
plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(history['val_iou'])
plt.title('Validation IoU Score')
plt.xlabel('Epoch')
plt.grid(True)

plt.tight_layout()
plt.show()
```
