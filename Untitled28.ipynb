```python
#assignment 1
def spiral_order(matrix):
    if not matrix:
        return

    top = 0
    bottom = len(matrix) - 1
    left = 0
    right = len(matrix[0]) - 1

    result = []

    while top <= bottom and left <= right:
        for j in range(left, right + 1):
            result.append(matrix[top][j])
        top += 1

        for i in range(top, bottom + 1):
            result.append(matrix[i][right])
        right -= 1

        if top <= bottom:
            for j in range(right, left - 1, -1):
                result.append(matrix[bottom][j])
            bottom -= 1

        if left <= right:
            for i in range(bottom, top - 1, -1):
                result.append(matrix[i][left])
            left += 1

    print(*result)


#assignment 2
def count_word_frequency(sentence):
    words = sentence.split()
    freq = {}

    for word in words:
        freq[word] = freq.get(word, 0) + 1

    return freq


#assignment 3
def is_palindrome_iterative(s):
    left = 0
    right = len(s) - 1

    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1

    return True

def is_palindrome_recursive(s):
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    return is_palindrome_recursive(s[1:-1])


#assignment 4
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False

    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2

    return True
```

```python
#assignment 1
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

normal_data = np.random.normal(0, 1, 1000)
beta_data = np.random.beta(2, 5, 1000)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].hist(normal_data, bins=20)
axes[0].set_title("Normal Distribution (mean=0, std=1)")
axes[0].set_xlabel("Value")
axes[0].set_ylabel("Frequency")

axes[1].hist(beta_data, bins=20)
axes[1].set_title("Beta Distribution (a=2, b=5)")
axes[1].set_xlabel("Value")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()


#aasignment 2
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

data = np.random.normal(0, 1, 1000)

mean = np.mean(data)
p25 = np.percentile(data, 25)
p75 = np.percentile(data, 75)

plt.figure(figsize=(8, 5))
plt.hist(data, bins=20, color="pink", edgecolor="black")

plt.axvline(mean, linestyle="--", label="Mean")
plt.axvline(p25, linestyle="--", label="25th Percentile")
plt.axvline(p75, linestyle="--", label="75th Percentile")

plt.title("Histogram of Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()

plt.show()
```

```python
#assignment 1
import numpy as np

np.random.seed(0)

arr = np.random.rand(3, 6)
add_arr = np.array([0, 0, 2, 4, 5, 3])

arr = arr + add_arr
reshaped = arr.reshape(9, 2)
transposed = reshaped.T

mean_val = np.mean(transposed)
locations = np.where(transposed > mean_val)

print("Final array:\n", transposed)
print("Mean:", mean_val)
print("Locations where elements > mean:", locations)


#assignment 2
import numpy as np

np.random.seed(0)

poisson_arr = np.random.poisson(lam=5, size=20)

mean_val = np.mean(poisson_arr)
std_val = np.std(poisson_arr)

normalized_arr = (poisson_arr - mean_val) / std_val

print("Original array:", poisson_arr)
print("Mean:", mean_val)
print("Standard Deviation:", std_val)
print("Centered and normalized array:", normalized_arr)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("videogamesales.csv")

df["global_sales"] = (
    df["NA_Sales"] +
    df["EU_Sales"] +
    df["JP_Sales"] +
    df["Other_Sales"]
)

df = df.sort_values(by="global_sales", ascending=False)
print(df.head())

genre_sales = df.groupby("Genre")["global_sales"].sum()

plt.figure(figsize=(10, 5))
genre_sales.plot(kind="bar")
plt.title("Total Global Sales by Genre")
plt.xlabel("Genre")
plt.ylabel("Copies Sold (Millions)")
plt.show()

gta_df = df[df["Name"].str.contains("Grand Theft Auto", case=False, na=False)]

gta_display = gta_df[[
    "Name",
    "Platform",
    "Year",
    "EU_Sales",
    "JP_Sales"
]]

gta_display["EU_JP_Sales"] = gta_display["EU_Sales"] + gta_display["JP_Sales"]

print(gta_display[["Name", "Platform", "Year", "EU_JP_Sales"]])

gta_region_sales = [
    gta_df["NA_Sales"].sum(),
    gta_df["EU_Sales"].sum(),
    gta_df["JP_Sales"].sum(),
    gta_df["Other_Sales"].sum()
]

labels = ["North America", "Europe", "Japan", "Other"]

plt.figure(figsize=(6, 6))
plt.pie(gta_region_sales, labels=labels, autopct="%1.1f%%", startangle=140)
plt.title("Regional Sales Distribution of Grand Theft Auto Games")
plt.show()
```
